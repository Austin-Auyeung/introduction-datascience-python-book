# Deep Learning

Artificial Neural Networks (ANN) is a biologically inspired programming paradigm which enables a computer to learn from observational data. The origins of neural networks can be traced, at least, until 1949, when D.Hebb proposed a simple learning principle for neurons: 'fire together, wire together'. Later, in 1957, F.Rosenblatt invented the first learning machine: the perceptron. These simple models were further developed for more than 30 years until 1988, when the backpropagation algorithm was proposed by Rumelhart et al. This algorithm was the first computational approach for training modern multilayer neural networks.

In spite of the fact that the discovery of the backpropagation algorithm started a rich research field leading to interesting models such as convolutional neural networks or recurrent neural networks, its severe practical limitations at that time provoked after a few years what is known as the neural net winter, a period of reduced funding and interest in the field. The neural net winter ended in 2012, when a neural network called AlexNet competed in the ImageNet Large Scale Visual Recognition Challenge. The network achieved a top-5 error of 15.3\%, more than 10.8 percentage points ahead of the runner-up. These results, which were unexpectedly good, led to the current deep learning boom.

Deep Learning is a powerful set of techniques (and tricks) for learning in neural networks with a high number of layers. Its notoriety builds upon the fact that it currently provides the best solutions to many problems in image recognition, speech recognition, and natural language processing, but at the same time some critics describe deep learning as a black box where data goes in and predictions come out without any kind of transparency.

# Neural Networks

Neural nets are a specific method for learning from data, a method that is based on a very simple element, the *neuron unit*. A neuron unit (or 1-layer neural network) is a mathematical function of this kind:

$${\mathbf y} = \sigma(\mathbf{w}^T \cdot {\mathbf x} + b)$$

where ${\mathbf x}$ represents an input element in vector form, $\mathbf{w}$ is a vector of weights,  $\sigma$ is a non-linear function and $b$ a scalar value. $(\mathbf{w},b)$ are called the parameters of the function. The output of this function is called the *activation* of the neuron. 

Regarding the non-linear function, historically the most common one was the Sigmoid function, but nowadays there are several alternatives that are supposed to be better suited to learning from data, such as ReLU and variants.

# Perceptron
The perceptron is a simple model which, given an input vector ${\mathbf x}$ of $m$ values $$(x_1, x_2, ..., x_n)$$ often called input features or simply features, outputs either 1 (yes) or 0 (no). Mathematically, the model is defined as:

$${\mathbf y} =  {\mathbf \sigma} \Big( \mathbf {w}  \cdot  {\mathbf x} + { b} \Big)$$

where $\sigma$ is known as the activation function, which in the case of the original perceptron is the linear threshold unit defined as:

$${\displaystyle f(\mathbf {x} )={\begin{cases}1&{\text{if }}\ \mathbf {w} \cdot \mathbf {x} +b>0,\\0&{\text{otherwise}}\end{cases}}}
$$

 As you can note, the Perceptron is really similar to Logistic Regression, but instead of providing a probabilistic output, it provides an output based on a hard threshold, the reason why Logistic Regression is usually preferred over Perceptron.
 
 
# Multi-Layer Perceptron
Simple neurons, such us Perceptrons, can be organized in larger structures by applying to the same data vector different sets of weights, forming what is called a *layer*, and by stacking layers one on top of the output of the other.  

It is important to notice that a multilayer neural network can be seen as a composition of matrix products (matrices represent weights) and non-linear activation functions. For the case of a 2-layer network, the outcome is:

$$ {\mathbf y} = {\mathbf \sigma}\Big( W^1  {\mathbf \sigma}\Big( W^0  {\mathbf x} + {\mathbf b}^0 \Big) + {\mathbf b}^1 \Big)$$

where ${\mathbf \sigma}$ represents a vectorial version of the sigmoid function and $W^i$ are the weights of each layer in matrix form.  

What is interesting about this kind of structure is that it has been shown that even a neural network with a single hidden layer containing a finite number of neurons can approximate any continuous function of $\mathbf{R}^n$. This fact makes neural networks a sound candidate to implement learning from data methods. The question is then: how to find the optimal parameters, ${\mathbf w} = (W^i,{\mathbf b})$, to approximate a function that is implicitly defined by a set of samples $\{({\mathbf x}_1, {\mathbf y}_1), \dots,  ({\mathbf x}_n, {\mathbf y}_n)\}$?

From a technical point of view, not only neural networks but most of the algorithms that have been proposed to infer models from large data sets are based on the iterative solution of a mathematical problem that involves data and a mathematical model. If there was an analytic solution to the problem, this should be the adopted one, but this is not the case for most of the cases. The techniques that have been designed to tackle these problems are grouped under a field that is called optimization. The most important technique for solving optimization problems is *gradient descend*.

> The training of models like $ {\mathbf y} = {\mathbf \sigma}\Big( W^1  {\mathbf \sigma}\Big( W^0  {\mathbf x} + {\mathbf b}^0 \Big) + {\mathbf b}^1 \Big)$ (or bigger!) can be readily performed by applying *automatic differentiation* to a loss function. 

> In the case of regression: $L = \frac{1}{n} \sum_{i=1}^n \Big({\mathbf y}_i - {\mathbf \sigma}\Big( W^1  {\mathbf \sigma}\Big( W^0  {\mathbf x}_i + {\mathbf b}^0 \Big) + {\mathbf b}^1 )\Big)\Big)^2 $

> In the case of two-class classification: $L = \frac{1}{n} log(1 + exp(-y_i {\mathbf \sigma}\Big( W^1  {\mathbf \sigma}\Big( W^0  {\mathbf x} + {\mathbf b}^0 \Big) + {\mathbf b}^1 \Big))) $

## Training neural networks

Training neural networks involves various methods, with one of the most common and simplest being the backpropagation algorithm. This method, published by Paul Werbos in 1982, was not entirely new; the algorithm he presented had already been discovered in the 1960s. In fact, Werbos himself mentioned this method in his PhD thesis in 1970, highlighting its potential for neural network learning.

Backpropagation is based on the chain rule and entails iteratively adjusting the weight connections to minimize a predefined loss function. This loss function quantifies the difference between the actual values and the values estimated by the network.

We will start with a simple dataset: the concentric circles.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split

X, y = make_circles(n_samples=1000, 
                            factor=.4, 
                            noise=0.05)

# Let us visualize the dataset
reds = y == 0
blues = y == 1
plt.figure()
plt.scatter(X[reds, 0], X[reds, 1], 
            c="red", s=20, edgecolor='k')
plt.scatter(X[blues, 0], X[blues, 1],
            c="blue", s=20, edgecolor='k')
plt.show()
```

Let us divide the data into train and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

The core data structure of Keras is a model, a way to organize layers. The main type of model is the ``Sequential model``, a linear stack of layers. 

Creating a MLP with two hidden layers is as straightforward as follow:

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=64, input_dim=2, activation="relu"))
model.add(Dense(units=64, activation="relu"))
model.add(Dense(units=1, activation="sigmoid"))
```

Here is a breakdown of the code:
* The first line initializes the Sequential model.
* The second line defines the first hidden layer with 64 units. Since this is the initial layer, we specify the number of input features (input\_dim=2) for our data, and we use the ReLU activation function.
* Next, we add a second hidden layer with 64 units, also using the ReLU activation function.
* Finally, we add the output layer with 1 unit, employing the sigmoid activation function. In this problem, the sigmoid function is suitable as it allows us to output values in the range [0, 1], making it suitable for binary classification tasks.

To obtain a quick overview of our model’s architecture we can call `model.summary()`. This function displays useful information about the layers, their types, output shapes, and the number of parameters. 

```python
model.summary()
```

As you can see, the first hidden layer consists of 192 parameters, while the second layer contains 4,160 parameters. These parameter counts are calculated as follows: for the first hidden layer, there are 2x64 connection weights plus an additional 64 biases, and for the second hidden layer, there are 64x64 connections plus an additional 64 biases

Once your model looks good, configure its learning process with ``.compile()``:

```python
model.compile(loss='binary_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

Here we are defining our loss function and the optimization method. The optimizer is used to tune the parameters of the neural network to minimize the loss function. In keras we have several optimizers methods, from the classical Stochastic Gradient Descend (SGD) to adaptative models such as RMSprop, Adam or Adagrad.

With your model compiled, you can now proceed to train it on your data in batches:

```python
model.fit(X_train, y_train,
          batch_size=32,
          epochs=5,
          verbose=1)
```

We can evaluate the model's performance on the test set using the following code:

```python
loss_and_metrics = model.evaluate(X_test, y_test)
```

Additionally, we can obtain predictions or probability scores for new data using the following code:

```python
classes = model.predict(X_test)
proba = model.predict_proba(X_test)
for i in range(5):
    print("predicted class:",classes[i][0],
          " -- prob(y==1):",proba[i][0])
    
```

We load the dataset and the corresponding training, validation and test set

```python
import os
import numpy as np
dataset_name = 'bloodmnist'
data = np.load(os.path.join(dataset.root, dataset_name + '.npz'))

X_train, y_train= data['train_images'],data['train_labels']
X_val, y_val    = data['val_images'],  data['val_labels']
X_test, y_test  = data['test_images'], data['test_labels']
```

After loading the data, you can check some information about the dataset,
such as the number of training examples and the dimensions of each sample

```python
num_train, height, width, channels  = X_train.shape 
print("There are", num_train,"training examples  ")
print("Each sample has a size of", height,"by", width,"pixels")

num_val  = X_val.shape[0] # there are 1712 test examples 
num_test = X_test.shape[0] # there are 3421 test examples
```

Following this, you can proceed to determine the number of unique classes
in the dataset

```python
num_classes = np.unique(y_train).shape[0]
```

To ensure proper data handling, we also normalize the data:

```python
X_train = X_train.astype('float32') 
X_val = X_val.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255.0 # Normalise data to [0, 1] range
X_val   /= 255.0 # Normalise data to [0, 1] range
X_test  /= 255.0 # Normalise data to [0, 1] range
```

Finally, we one-hot encode the labels for training, validation and test sets:

```python
from keras.utils import np_utils

y_train = np_utils.to_categorical(y_train, num_classes) 
y_val = np_utils.to_categorical(y_val, num_classes) 
y_test = np_utils.to_categorical(y_test, num_classes) 
```

With these steps, the data is correctly loaded, prepared, and ready for further processing. To visualize some samples, we can use the following code:

```python
plt.figure(figsize=(16,12))
for i in range(24):
    plt.subplot(4,6,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i].reshape((height,width,channels)), cmap=plt.cm.binary)
    plt.xlabel(dataset.info['label'][str(np.argmax(y_train[i]))][:20])
plt.show()
```

We are now ready to create our first model. Let us create a simple MLP model. In the following example, we create a MLP model with two hidden layers with 256 neurons each and one output layer with as many neurons as the number of different categories in the dataset.

```python
from keras.layers import Flatten

## DEFINE PARAMETERS
batch_size = 128
epochs = 20

## CREATE THE MODEL
model = Sequential()
model.add(Flatten(input_shape=[height, width, channels]))
model.add(Dense(32, activation='relu', input_shape=(height*width*channels,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

We can check the model summary to see its architecture and the number of parameters:

```python
model.summary()
```

As it  can see, the created models has 75.960 paraments that needs to be optimized. That's a big number. 
In order to optimize all this parameters we need to define the loss function and the optimizer method. 

```python
## COMPILE MODEL -> loss + optimizer
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

Now, we are ready to train it!!

```python
history_mlp = model.fit(X_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(X_val, y_val))
```

When plotting the learning curves we can see that after 20 epochs training and validation are still improving without any overfitting signal.

```python
def plot_history(histories):
    fig, ax = plt.subplots(1, 2,figsize=(15, 7))
    for (history,name) in histories:
        ax[0].plot(history.history['accuracy'],'o-.',label  = name+' train')
        ax[0].plot(history.history['val_accuracy'],'x-',label  = name+' test')

        ax[0].set_title('model accuracy')
        ax[0].set_ylabel('accuracy')
        ax[0].set_xlabel('epoch')
        ax[0].legend(loc='upper left')

    for (history,name) in histories:
        ax[1].plot(history.history['loss'],'o-.',label  = name+' train')
        ax[1].plot(history.history['val_loss'],'x-',label  = name+' test')

        ax[1].set_title('model loss')
        ax[1].set_ylabel('loss')
        ax[1].set_xlabel('epoch')
        ax[1].legend(loc='upper left')
    plt.show()
    
plot_history([(history_mlp,'MLP')])
```

Finally, we evaluate the model on the test set, and the results should be similar to the validation set:

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

# Convolution Neural Networks 
While multilayer perceptrons (MLPs) are powerful neural network models, they have limitations when it comes to processing raw image data. MLPs are structured in layers, with each neuron in a layer receiving input from all neurons in the previous layer. This design is suitable for certain problems, particularly those with a fixed number of relatively unstructured parameters.

However, let's consider what happens when we feed raw image data, such as a 200x200 pixel image, into an MLP with 1024 neurons:
> The first hidden layer alone would require more than 40 million parameters.

The situation quickly becomes unmanageable as image sizes grow larger, way before reaching the kind of images people usually want to work with in real applications.

To address this challenge, a common solution is to downsample images to a size where MLPs can be applied more effectively. However, this downsampling process risks losing valuable information. Ideally, we'd like to perform useful image processing before downsampling without significantly increasing the number of parameters.

Fortunately, Convolutional Neural Networks (CNNs) offer an efficient solution. CNNs take advantage of the inherent structure in images. They assume that pixels spatially close to each other collaborate more closely to form specific features of interest, compared to pixels at opposite corners of the image. Additionally, if a particular feature is essential for defining an image's label, its presence anywhere within the image should be equally important, regardless of its location. CNNs leverage these principles to effectively process and extract features from images, making them a key tool in image analysis and recognition tasks.


At the core of the CNNs lies the convolution operation. Given a two-dimensional image, $I$, and a small matrix, $K$ (known as a convolution kernel) of size $h \times w$, which encodes a method for extracting image features, we compute the convolved image, $I∗K$, by overlaying the kernel on the image in all possible positions and recording the sum of element-wise products between the image and the kernel:

$$
output(x,y) = (I \otimes K)(x,y) = \sum_{m=0}^{M-1} \sum_{n=1}^{N-1} K(m,n) I(x-n, y-m)
$$

This convolution operator forms the fundamental basis of the convolutional layer of a CNN. The layer is completely specified by a certain number of kernels, $K$, and it operates by computing the convolution of the output images of a previous layer with each of those kernels, afterwards adding the biases (one per each output image). Finally, an activation function, $\sigma$, may be applied to all of the pixels of the output images. 

Typically, the input to a convolutional layer will have $d$ channels (e.g., red/green/blue in the input layer), in which case the kernels are extended to have this number of channels as well.

Note that, since all we're doing here is addition and scaling of the input pixels, the kernels may be learned from a given training dataset via gradient descent, exactly as the weights of an MLP. In fact, an MLP is perfectly capable of replicating a convolutional layer, but it would require a lot more training time (and data) to learn to approximate that mode of operation.

### Building Blocks of a CNN

Now, let's explore the main components that make up a CNN:


Input Layer: Although CNNs are versatile and not limited to images, they are commonly used for image data, which is typically represented as $h \times w \times d$ ($h$ = Height, $w$ = Width, $d$ = Channels). This is the first layer that receives the input data.


**Input Layer**

Although CNNs are versatile and not limited to images, they are commonly used for image data, which is typically represented as $h \times w \times d$ ($h$ = Height, $w$ = Width, $d$ = Channels). This is the first layer that receives the input data.

**Convolutional Layer**

This is the heart of a CNN and often consists of multiple layers in the network. These are the layers that will perform most of the architecture's computational work.

**Sub-sampling or Pooling Layer**

This layer serves a dual purpose. First, it creates a smoother representation, reducing sensitivity to noise and variations in the data. Second, it reduces the size of the representation, which lowers the number of parameters and system complexity, helping to prevent overfitting. Sub-sampling typically involves resizing the input to a smaller output size, controlled by the stride parameter. A common approach is to use max-pooling with a filter size of 2x2 (or 4x4) and a stride of two pixels.

```python
model = Sequential()
model.add(Conv2D(filters = 32, kernel_size = (3,3), 
                 activation = 'relu', 
                 input_shape = (height,width,channels)))

model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(filters = 16, kernel_size = (3,3), 
                 activation = 'relu'))
model.add(Flatten())
model.add(Dense(8,activation='softmax'))
```

This output size is derived from our input images, which are of size $28𝑥28𝑥3$, accounting for the three color channels (r, g, b). Notably, the filters in this layer are sized at $3𝑥3𝑥3$ to match the input data's three channels (r, g, b).

* **Pooling Layer:** Following the convolutional layer, there is a pooling layer with a pool size of 2, resulting in a halving of the input's dimensions. Therefore, the output shape becomes $13x13x32$.

* **Second Convolutional Layer:** This layer features 16 filters, each with dimensions of $3x3x32$. Consequently, the output shape is $11x11x16$.

* **Output Layer:** Finally, we have the output layer, which is a dense layer comprising 8 neurons, matching the number of categories in our problem. Note that we previously flattened the input data since the dense layer requires a 1D array.

## Avoiding overfitting
With milions of parameters we can fit anything. But the great flexivity that offers this amout of parameters make the network prone to overfit the training data. In oder to avoid there is some popular techiniques such as: 1) Dropout; 2) $l_1$ and $l_2$ regularization; or 3) Data Augmentation.

### Dropout

Dropout is a valuable technique for regularizing neural networks. During training, it's possible for neurons in a particular layer to become overly dependent on the output of specific neurons in the preceding layer. This situation can lead to overfitting.

Dropout mitigates the risk of overfitting by introducing randomness. It does so by intermittently severing connections, or "dropping" connections, between neurons in successive layers during training. This randomness helps ensure that no single neuron dominates the learning process, promoting a more robust and generalized model.

Below we can see how dropout can be used in keras:

```python
from keras.layers import Dropout

model = Sequential()
model.add(Conv2D(256, (3,3), activation = 'relu',
                 input_shape = (height,width,channels)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(num_classes, activation='softmax'))
```

Convolutional layers apply a local receptive field over the input data, which forces weight sharing among neurons. This weight sharing reduces the risk of overfitting by limiting the number of independent parameters in the network, making explicit regularization techniques less necessary.

### Data Augmentation
Another way to prevent overfitting is by means of data augmentation. Data augmentation increases the size of your training dataset by applying various transformations to the existing images. These transformations include:

* Rotation: Rotating the image by a certain angle, which can help the model become more invariant to the orientation of objects.

* Width and Height Shifts: Shifting the image horizontally or vertically, simulating variations in object positioning within the frame.

* Shear: Distorting the image by shifting one part of the image relative to the other along a given axis. This can simulate the effect of tilting or slanting an object.

* Zoom: Zooming in or out of the image, changing the scale and perspective.

* Horizontal Flip: Flipping the image horizontally, which is useful for tasks where left-to-right orientation doesn't matter.

By applying these transformations randomly to your training images, you effectively create new training examples, each slightly different from the original. This helps the model generalize better because it learns to recognize objects under various conditions and viewpoints.

In the next code the ImageDataGenerator from TensorFlow/Keras is used to automate the process of data augmentation. It takes care of applying these transformations to the training images on-the-fly as the model is trained, effectively expanding the training dataset and enhancing the model's ability to generalize to unseen data.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the data augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')


model = Sequential()
model.add(Conv2D(256, (3,3), activation = 'relu',
                 input_shape = (height,width,channels)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Fit the model with data augmentation
history = model.fit(datagen.flow(X_train, y_train, batch_size=batch_size),
                    steps_per_epoch=len(X_train) / batch_size,
                    epochs=epochs,
                    validation_data=(X_test, y_test))
```
